
Write a function that checks if customer borrowed more or less films in the month of June as compared to May.

Hint: For this part, you can create a join between the two dataframes created before, using the merge function available for pandas dataframes. 
Here is a link to the documentation for the merge function.


```python
merge = df_3.merge(df_1, how="inner", on="customer_id", suffixes=('_juin', '_may'))
merge["différence"] = merge["number_rental_may"] - merge["number_rental_juin"] 

merge
```

```python
merge["différence"] = merge["différence"].astype(int)
```

```python
def film_rental(x):
    if x == 0: 
        return "no winner"
    elif x > 0:
        return "may" 
    else:
        return "june"
```

```python
def film_1(june, may):
    if june == may: 
        return "no winner"
    elif may > june:
        return "may" 
    else:
        return "june"
```

```python
merge["winner_1"] = merge.apply(lambda x: film_1(x.number_rental_juin, x.number_rental_may), axis=1)
```
